```python
import tensorflow as tf
import numpy as np
```

```python
def load(file_path):
  with open(file_path, 'r') as f:
    data = f.readlines()
  return data
file_path="Harry-Potter-1.txt"
text = load(file_path)
```

```python
text
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(oov_token='<OOV>')
tokenizer.fit_on_texts([text])
print(tokenizer.word_index)
total_words=len(tokenizer.word_index)+1
print(total_words)
```

```python
input_sequences = []
tokens = tokenizer.texts_to_sequences([text])[0]
print(tokens)
```

```python
sq_len =50
for i in range(sq_len , len(tokens)):
  input_sequences.append(tokens[i-sq_len:i+1])
```

```python
len(input_sequences[2])
```

```python
from tensorflow.keras.utils import pad_sequences,to_categorical
```

```python
from re import X
input_sequences = np.array(pad_sequences(input_sequences,maxlen=sq_len+1,padding='pre'))
X = input_sequences[:,:-1]
y = input_sequences[:,-1]
```

```python
X
```

```python
y=tf.keras.utils.to_categorical(y,num_classes=total_words)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN,Embedding,Dense
```

```python
model = Sequential([
    Embedding(input_dim = total_words,output_dim = 64,input_length=sq_len),
    SimpleRNN(256),
    Dense(256,activation='relu'),
    Dense(total_words,activation='softmax')
]
)
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(X,y,validation_split=0.2,epochs = 10,batch_size = 120)
```
